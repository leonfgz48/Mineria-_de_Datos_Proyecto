```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from datetime import datetime
import ast
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# DATOS CREADOS POR EMMANUEL

categoricas = pd.read_csv('../data/variables_categoricas_airbnb.csv')
categoricas['id'] = categoricas['id'].astype('int')
categoricas = categoricas.set_index('id')
```

```python
# DATOS ORIGINALES

listings = pd.read_csv('../data/raw_list.csv')
listings['id'] = listings['id'].astype('int')
listings = listings.set_index('id')
```

```python
listings.columns.sort_values()
```

```python
listings[['number_of_reviews']]
```

```python
num_cols = ['host_total_listings_count',
            'host_acceptance_rate',
            'host_response_rate',
            'calculated_host_listings_count_entire_homes',
            'calculated_host_listings_count_private_rooms',
            'calculated_host_listings_count_shared_rooms',
            'latitude',
            'longitude',
            'accommodates',
            'bedrooms',
            'beds',
            'price']
```

```python
listings_num = listings[num_cols]
```

```python
listings_num.corr()
```

# Limpieza de columnas en texto

```python
listings_num['host_acceptance_rate'] = listings_num['host_acceptance_rate'].str.replace('%', '')\
                                       .astype(float) / 100

listings_num['host_response_rate'] = listings_num['host_response_rate'].str.replace('%', '')\
                                     .astype(float) / 100
```

```python
listings_num['price'] = listings_num['price'].str.replace('$', '').str.replace(',', '').astype(float)
```

# Concatenación de tablas

```python
categoricas = categoricas.reset_index().drop(['id'], axis = 1)
listings_num = listings_num.reset_index().drop(['id'], axis = 1)
```

```python
listints_concat = pd.concat([categoricas, listings_num], axis = 1)
```

# Valores Nulos

```python
listints_concat.isnull().sum()
```

```python
listints_concat['bedrooms'].fillna(0, inplace=True)
listints_concat['beds'].fillna(1, inplace=True)
```

```python
listints_concat['host_acceptance_rate'].fillna(listints_concat['host_acceptance_rate'].mean(), inplace=True)
listints_concat['host_response_rate'].fillna(listints_concat['host_response_rate'].mean(), inplace=True)
```

```python
listints_concat = listints_concat.dropna(subset=['price'])
```

```python
listints_concat.isnull().sum()
```

```python
listints_concat
```

# Manejo de Valores Atípicos

```python
Q1 = listints_concat['host_total_listings_count'].quantile(0.25)
Q3 = listints_concat['host_total_listings_count'].quantile(0.75)
IQR = Q3 - Q1
listints_concat_clean = listints_concat[~((listints_concat['host_total_listings_count'] < (Q1 - 1.5 * IQR)) | 
                                         (listints_concat['host_total_listings_count'] > (Q3 + 1.5 * IQR)))]
```

```python
quantile_99 = listints_concat_clean['price'].quantile(0.999)

listints_concat_clean = listints_concat_clean[listints_concat_clean['price'] <= quantile_99]
```

# Generación de nuevos archivos

```python
listints_concat.to_csv('../data/eda_data/variables_con_atípicos.csv')
listints_concat_clean.to_csv('../data/eda_data/variables_sin_atípicos.csv')
```
